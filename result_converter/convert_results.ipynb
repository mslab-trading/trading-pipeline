```python
import os

from RegressionResult import RegressionResult
```

```python
# Example of converting from pipeline format to model format
input_path_root = "../results/Example_Result/"
output_path_root = "results/model_output/"

for dir in os.listdir(input_path_root):
    input_path = os.path.join(input_path_root, dir)
    output_path = os.path.join(output_path_root, dir)
    if not os.path.isdir(input_path):
        continue

    result = RegressionResult.from_pipeline(input_path)
    result.to_model(output_path)
    print(f"Converted {input_path} to {output_path}")
```

```python
# Example of transforming from model format to pipeline format
input_paths = [
    f"/data2/b09902053/trading-model/results/LSTM_{year-3}_{year-1}_Semiconductor_max_roi_ConcordanceCorrelation_30_60_0_broker"
    for year in range(2021, 2025)
]
output_path_root = "results/pipeline_output/"

for input_path in input_paths:
    dir = os.path.basename(input_path)
    output_path = os.path.join(output_path_root, dir)
    result = RegressionResult.from_model(input_path)
    result.to_pipeline(output_path)
    print(f"Converted {input_path} to {output_path}")
```
